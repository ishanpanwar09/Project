```python
# Dependencies
import os
import pandas as pd
import nltk
import gensim
from gensim import corpora, models, similarities
```

```python
# Uncomment if want to work on all classes!
concert = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Concert_metadata.csv').iloc[:,1:]
#conference = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Conference_metadata.csv').iloc[:,1:]
#cycling = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Cycling_metadata.csv').iloc[:,1:]
#football = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Football_metadata.csv').iloc[:,1:]
#frames = [concert, conference, cycling, football]
# df = pd.concat(frames)

# comment this DF and uncomment the one above to use data for all class
df = pd.DataFrame(concert)
```

```python
# cc resembles class count and then forms a corpus or bag of words or just unique word grouped by count
cc = df.groupby('name').count()
cc['name'] = cc.index
cc = cc[['name','value']]
corpus = cc.name.tolist()
corp = []
for i in corpus:
    corp.append(i.replace(' ',''))
corpus = corp   
```

```python
tok_corp= [nltk.word_tokenize(sent) for sent in corpus]           
model = gensim.models.Word2Vec(tok_corp, min_count=1, size = 2)
# save the model for re-use
#model.save('similatConcepts_ait582_isingh3')
#model = gensim.models.Word2Vec.load('similatConcepts_ait582_isingh3')
```

```python
# Extraction of similar keywords
x = []
name = []
for i in tok_corp:
    a = list(enumerate(model[i]))
    x.append(a[0][1])
    name.append(i)

vec= pd.DataFrame(x)
vec['names'] = name
vec.columns = ['x','y','word']
vec = vec[['word','x','y']]
```

```python
points = vec
def plot_region(x_bounds, y_bounds):
    slice = vec[
        (x_bounds[0] <= vec.x) &
        (vec.x <= x_bounds[1]) & 
        (y_bounds[0] <= vec.y) &
        (vec.y <= y_bounds[1])
    ]
    
    ax = slice.plot.scatter("x", "y", s=40, figsize=(10, 8))
    for i, vec in slice.iterrows():
        ax.text(vec.x + 0.005, vec.y + 0.005, vec.word, fontsize=11)
```

```python
# removing idiotism from Words
w = []
for i in vec.word:
    w.append(i[0])
vec['newWord'] = w
vec = vec[['word', 'newWord','x','y']]
vec = vec.drop('word', axis=1)
```
