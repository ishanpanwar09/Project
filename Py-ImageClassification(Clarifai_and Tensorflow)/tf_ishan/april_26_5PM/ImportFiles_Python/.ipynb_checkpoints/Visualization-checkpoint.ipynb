```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# concert
#pdf = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Concert_metadata.csv')
# 
#conference_metadata
#pdf = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/conference_metadata.csv')
#Cycling_metadata
#pdf = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Cycling_metadata.csv')
#Football_metadata
pdf = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Football_metadata.csv')
```

```python
pdf = pdf.groupby('name').count()

# Important MetaData
imp20Metadata_Concert = pdf[pdf.value >= 30]
lessImpMetadata_Concert = pdf[pdf.value < 30]
```

```python
a = pd.DataFrame(imp20Metadata_Concert)
a['name'] = a.index
a.columns = ['id', 'value','name']

# select particular highest
# best5 = a.nlargest(5,'value')

sns.set(style="ticks", color_codes=False)
g=sns.barplot(x="name", y="value", data=a)
plt.setp(g.get_xticklabels(), rotation=65)
plt.suptitle('Most Important Concepts Extracted by Frequency for Football Class')
g.set(xlabel='Metadata Concepts', ylabel='Frequency')
plt.show()
```
