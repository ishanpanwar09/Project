```python
import tensorflow as tf, sys
import glob

# reading all files for testing
path_f1 = r'/Users/ishansingh/Documents/tf_ishan/Tensorflow_Classification/tf_files/test_data/concert'
allFiles1 = glob.glob(path_f1 + "/*.jpg")
path_f2 = r'/Users/ishansingh/Documents/tf_ishan/Tensorflow_Classification/tf_files/test_data/conference'
allFiles2 = glob.glob(path_f2 + "/*.jpg")
path_f3 = r'/Users/ishansingh/Documents/tf_ishan/Tensorflow_Classification/tf_files/test_data/football'
allFiles3 = glob.glob(path_f3 + "/*.jpg")
path_f4 = r'/Users/ishansingh/Documents/tf_ishan/Tensorflow_Classification/tf_files/test_data/cycling'
allFiles4 = glob.glob(path_f4 + "/*.jpg")

# combining all files in different folders to put in single loop
allFiles = allFiles1 + allFiles2 + allFiles3 + allFiles4

# extracting actual class
actual = []
for i in range(0,len(allFiles)):
    a = allFiles[i].split('/')[8]
    actual.append(a)


file = [] # for file names
cl = []  # for class names
sc = []  # for score of the predicted classes
ccl = [] # for score of the actual classes

for i in range(0,len(allFiles)):
    #looping over the list of file paths
    image_path = allFiles[i]
    # Read in the image_data
    image_data = tf.gfile.FastGFile(image_path, 'rb').read()

    # Loads label file, strips off carriage return
    label_lines = [line.rstrip() for line
                   in tf.gfile.GFile("/Users/ishansingh/Documents/tf_ishan/Tensorflow_Classification/tf_files/retrained_labels.txt")]

    # Unpersists graph from file
    with tf.gfile.FastGFile("/Users/ishansingh/Documents/tf_ishan/Tensorflow_Classification/tf_files/retrained_graph.pb", 'rb') as f:
        graph_def = tf.GraphDef()
        graph_def.ParseFromString(f.read())
        _ = tf.import_graph_def(graph_def, name='')
    
    with tf.Session() as sess:
    
        
    # Feed the image_data as input to the graph and get first prediction
        softmax_tensor = sess.graph.get_tensor_by_name('final_result:0')
    
        predictions = sess.run(softmax_tensor, \
             {'DecodeJpeg/contents:0': image_data})
    
        # Sort to show labels of first prediction in order of confidence
        top_k = predictions[0].argsort()[-len(predictions[0]):][::-1]
    
        for node_id in top_k:
            human_string = label_lines[node_id]
            score = predictions[0][node_id]
            file.append(allFiles[i])
            ccl.append(actual[i])
            cl.append(human_string)
            sc.append(score)
            print('%s  ||  %s  ||  %s  (score = %.5f)' % (allFiles[i],i,human_string, score))
```

```python
import pandas as pd
newFile = [] 
newClass = [] 
newScore = []
act_class = []
newFile = newFile + file
newClass = newClass + cl
newScore = newScore + sc
act_class = act_class + ccl

Results = pd.DataFrame()
Results['FileName'] = newFile
Results['Actual_Class'] = act_class
Results['Predicted_Class'] = newClass
Results['Score'] = newScore
```
