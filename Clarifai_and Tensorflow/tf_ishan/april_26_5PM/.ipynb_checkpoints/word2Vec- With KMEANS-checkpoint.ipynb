```python
# the semicolon YOUTUBE
import os
import pandas as pd
import nltk
import gensim
from gensim import corpora, models, similarities
```

```python
# reading Files
concert = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Concert_metadata.csv').iloc[:,1:]
conference = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Conference_metadata.csv').iloc[:,1:]
cycling = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Cycling_metadata.csv').iloc[:,1:]
football = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Football_metadata.csv').iloc[:,1:]
frames = [concert, conference, cycling, football]
df = pd.concat(frames)
```

```python
# cleaning and Arranging data for further feeding 
co = df.name.tolist()
corpus = co
corp = []
for i in corpus:
    corp.append(i.replace(' ',''))
corpus = corp 
```

```python
# vectorizing words
tok_corp= [nltk.word_tokenize(sent) for sent in corpus]           
model = gensim.models.Word2Vec(tok_corp, min_count=1, size = 300)
```

```python
# ETL for proper access later
x = []
name = []
for i in tok_corp:
    a = list(enumerate(model[i]))
    x.append(a[0][1])
    name.append(i)
# dataframe creation    
vec= pd.DataFrame(x)
```

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn import metrics

# transforming data by scaling and then applying T-SNE
standard_scaler = StandardScaler()
x_std = standard_scaler.fit_transform(vec)
tsne = TSNE(n_components=2, random_state=0)
x_test_2d = tsne.fit_transform(x_std)
df = pd.DataFrame(x_test_2d)
df.columns = [['x','y']]
```

```python
# Clustering 
kmeans = KMeans(init='k-means++', n_clusters=4, n_init=10)
kmeans.fit(df)
```

```python
import matplotlib.pyplot as plt
markers=('s', 'd', 'o', '^', 'v')
color_map = {0:'red', 1:'blue', 2:'lightgreen', 3:'purple', 4:'cyan'}
df.plot.scatter("x", "y", s=5, figsize=(10, 8))
plt.title('T-SNE reduced clustering with Kmeans')
plt.show()
```

```python
from time import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
from sklearn import metrics
kmeans = KMeans(init='k-means++', n_clusters=4, n_init=10, max_iter= 400, precompute_distances= True)
kmeans.fit(df)
```

```python
# Step size of the mesh. Decrease to increase the quality of the VQ.
h = .02     # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = df.x.min() - 1, df.x.max() + 1
y_min, y_max = df.y.min() - 1, df.y.max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
# Put the result into a color plot
Z = Z.reshape(xx.shape)
#plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired, aspect='auto', origin='lower')
plt.plot(df.x, df.y, 'k.', markersize=2)
# Plotting centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=169, linewidths=3, color='w', zorder=10)
plt.title('K-means clustering on the extracted metadata \n'
            'from Images (T-SNE-reduced data) \n'
          'Centroids are marked with white cross')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
