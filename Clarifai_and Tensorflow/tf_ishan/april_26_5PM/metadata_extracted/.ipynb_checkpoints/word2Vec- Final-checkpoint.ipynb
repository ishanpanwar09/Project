```python
#os.chdir("D:\semicolon\Deep Learning");
concert = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Concert_metadata.csv').iloc[:,1:]
conference = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Conference_metadata.csv').iloc[:,1:]
cycling = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Cycling_metadata.csv').iloc[:,1:]
football = pd.read_csv('/Users/ishansingh/Documents/tf_ishan/april_26_5PM/metadata_extracted/Football_metadata.csv').iloc[:,1:]
frames = [concert, conference, cycling, football]
df = pd.concat(frames)
```

```python
#cc = df.groupby('name').count()
#cc['name'] = cc.index
#cc = cc[['name','value']]
#corpus = cc.name.tolist()
co = df.name.tolist()
corpus = co
corp = []
for i in corpus:
    corp.append(i.replace(' ',''))
corpus = corp 
```

```python
tok_corp= [nltk.word_tokenize(sent) for sent in corpus]           
model = gensim.models.Word2Vec(tok_corp, min_count=1, size = 300)
#model.save('testmodel')
#model = gensim.models.Word2Vec.load('test_model')
#model.most_similar('word')
#model.most_similar([vector])
```

```python
#model = gensim.models.Word2Vec.load('test_model')
#model.most_similar('city')
#model.most_similar([vector])
# Extraction of similar keywords
x = []
name = []
for i in tok_corp:
    a = list(enumerate(model[i]))
    x.append(a[0][1])
    name.append(i)
```

```python
vec= pd.DataFrame(x)
```

```python
#a= list(enumerate(model['people']))
#a[0][1], a[1][1], a
#x = vec.ix[:,:-1].values
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
x_std = standard_scaler.fit_transform(vec)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
x_test_2d = tsne.fit_transform(x_std)
```

```python
df = pd.DataFrame(x_test_2d)
df.columns = [['x','y']]
```

```python
import matplotlib.pyplot as plt
markers=('s', 'd', 'o', '^', 'v')
color_map = {0:'red', 1:'blue', 2:'lightgreen', 3:'purple', 4:'cyan'}
df.plot.scatter("x", "y", s=5, figsize=(10, 8))
plt.title('T-SNE For reducing the 300 dimensions to a plotable 2 Dimensions')
plt.show()
```
